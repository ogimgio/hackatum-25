```python
import requests

BASE_URL = "https://hackatum25.sixt.io"

def create_booking():
    """
    Creates a booking and returns the booking ID.
    """
    url = f"{BASE_URL}/api/booking"
    response = requests.post(url)

    if response.status_code != 200:
        raise Exception(f"Create booking failed: {response.status_code} - {response.text}")

    data = response.json()
    print("Created booking:", data)
    return data["id"]


def get_available_vehicles(booking_id):
    """
    Retrieves available vehicles for a booking.
    """
    url = f"{BASE_URL}/api/booking/{booking_id}/vehicles"
    response = requests.get(url)

    if response.status_code != 200:
        raise Exception(f"Get vehicles failed: {response.status_code} - {response.text}")

    return response.json()


def main():
    # 1. Create booking
    booking_id = create_booking()

    # 2. Get available vehicles
    vehicles = get_available_vehicles(booking_id)

    # 3. Print available vehicles
    print("\nAvailable vehicles:")
    print(vehicles)

if __name__ == "__main__":
    main()
```
